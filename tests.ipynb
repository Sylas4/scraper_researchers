```python
def scrape_dynamic_content(url):
    
    # Set up Selenium with Chrome (headless mode optional)
    options = Options()
    options.add_argument("--headless")  # Run in headless mode
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.5615.121 Safari/537.36")

    service = Service('/usr/local/bin/chromedriver')
    driver = webdriver.Chrome(options=options, service=service)

    try:
        # Load the page
        driver.get(url)

        # Wait for the content to load (e.g., wait for a specific element)
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.rims-filter-group"))  # Replace with actual selector
        )

        # Get the fully rendered page source
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # Extract the desired content
        data = []
        for item in soup.select("div.rims-filter-group"):  # Update the selector based on the structure
            title = item.select_one("a").get_text(strip=True)
            link = item.select_one("a").get("href")
            data.append({"title" : title, "link" : link})

        return data

    finally:
        driver.quit()

if __name__ == "__main__":
    url = "https://ki.se/en/research/research-areas-centres-and-networks/research-groups/research-groups-in-anesthesiology-and-intensive-care"
    content = scrape_dynamic_content(url)
    for c in content:
        print(c)
```

```python
import requests 
from bs4 import BeautifulSoup 
import csv
import pandas as pd

def scrape_ehpads(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error requesting page: {response.status_code}")
        return []

    soup = BeautifulSoup(response.content, 'html.parser')
    ehpads = soup.select_one('div.view.annuaire-list.active').select('section.result-entry.h-card') 

    data = []
    for item in ehpads:  
        name = item.select_one('a.p-name') 
        phone = item.select_one('a.p-tel') 
        if name and phone:
            data.append({
                "name": name.get_text(strip=True),
                "phone": phone.get_text(strip=True)
            })

    return data

def save_to_csv(data, filename):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=["name", "phone"])
        writer.writeheader()
        writer.writerows(data)

def save_to_xls(data, filename):
    df = pd.DataFrame(data)
    df.to_excel(filename, index=False, engine='openpyxl')

if __name__ == "__main__":
    url = "https://www.pour-les-personnes-agees.gouv.fr/annuaire-ehpad-et-comparateur-de-prix-et-restes-a-charge/paris-75"
    
    data = scrape_ehpads(url)
    if data:
        save_to_csv(data, "ehpads_paris.csv")
        print("Data saved in  'ehpads_paris.csv'")
    else:
        print("No data found")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

class EhpadScraper:
    def __init__(self, url):
        self.url = url
        self.soup = None

    def fetch_page(self):
        """Fetch the HTML content of the page and parse it with BeautifulSoup."""
        response = requests.get(self.url)
        if response.status_code != 200:
            raise Exception(f"Error requesting page: {response.status_code}")
        self.soup = BeautifulSoup(response.content, 'html.parser')

    def scrape_data(self):
        """Scrape EHPAD data from the parsed HTML."""
        if self.soup is None:
            raise Exception("Soup object is not initialized. Call fetch_page() first.")

        ehpads = self.soup.select_one('div.view.annuaire-list.active').select('section.result-entry.h-card')
        data = []

        for item in ehpads:
            name = item.select_one('a.p-name')
            phone = item.select_one('a.p-tel')  # Assuming a 'phone' class exists for phone numbers
            if name and phone:
                data.append({
                    "name": name.get_text(strip=True),
                    "phone": phone.get_text(strip=True)
                })

        return data

    def save_to_csv(self, data, filename):
        """Save the scraped data to a CSV file."""
        with open(filename, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.DictWriter(file, fieldnames=["name", "phone"])
            writer.writeheader()
            writer.writerows(data)

    def save_to_excel(self, data, filename):
        """Save the scraped data to an Excel file."""
        df = pd.DataFrame(data)
        df.to_excel(filename, index=False, engine='openpyxl')


if __name__ == "__main__":
    # Initialize the scraper
    url = "https://www.pour-les-personnes-agees.gouv.fr/annuaire-ehpad-et-comparateur-de-prix-et-restes-a-charge/paris-75"
    scraper = EhpadScraper(url)

    try:
        # Fetch and scrape the data
        scraper.fetch_page()
        data = scraper.scrape_data()

        if data:
            # Save to CSV and Excel
            scraper.save_to_csv(data, "ehpads_paris.csv")
            scraper.save_to_excel(data, "ehpads_paris.xlsx")
            print("Data saved in 'ehpads_paris.csv' and 'ehpads_paris.xlsx'")
        else:
            print("No data found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
